```python
from bs4 import BeautifulSoup
import requests

url = 'https://results.eci.gov.in/'

response = requests.get(url)
response.raise_for_status()  

soup = BeautifulSoup(response.content, 'html.parser')

sub_headings = soup.find_all('h2')
main_headings = soup.find_all('h1')[1:]

for sub_heading, main_heading in zip(sub_headings, main_headings):
    print(f'{main_heading.text.strip()} - {sub_heading.text.strip()}')

max_len = max(len(sub_headings), len(main_headings))
for i in range(len(sub_headings), max_len):
    print(f'{main_headings[i].text.strip()} - ')
for i in range(len(main_headings), max_len):
    print(f' - {sub_headings[i].text.strip()}')
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'https://results.eci.gov.in/AcResultByeJune2024/'

response = requests.get(url)
response.raise_for_status() 

soup = BeautifulSoup(response.content, 'html.parser')

h2_headings = soup.find_all('h2')
h3_headings = soup.find_all('h3')
h4_headings = soup.find_all('h4')
h5_headings = soup.find_all('h5')
h6_headings = soup.find_all('h6')[1:][:-1]

max_len = max(len(h2_headings), len(h3_headings), len(h4_headings), len(h5_headings), len(h6_headings))

data = []

for i in range(max_len):
    h2_text = h2_headings[i].text.strip() if i < len(h2_headings) else ''
    h3_text = h3_headings[i].text.strip() if i < len(h3_headings) else ''
    h4_text = h4_headings[i].text.strip() if i < len(h4_headings) else ''
    h5_text = h5_headings[i].text.strip() if i < len(h5_headings) else ''
    h6_text = h6_headings[i].text.strip() if i < len(h6_headings) else ''
    
    data.append([h4_text, h3_text, h2_text, h5_text, h6_text])

print("Bye Election to Assembly Constituencies Results June-2024")
df = pd.DataFrame(data, columns=['State','Area','Status','MLA Name','Party'])

df
```
